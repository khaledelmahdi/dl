```python
# Import Library
import pandas as pd
```

```python
#Import the files to Google Colab
url = 'https://raw.githubusercontent.com/rc-dbe/bigdatacertification/master/dataset/churn_trasnsformed_new.csv'
df_csv = pd.read_csv(url, sep=',',)

# Show 10 first Row
df_csv.head()
```

```python
# Remove "Unnamed:O" Coloumn
df = df_csv.drop("Unnamed: 0", axis=1)
df.head()
```

```python
# Check the Data Information
df.info()
```

```python
#Import MinMax Scaler
from sklearn.preprocessing import MinMaxScaler

# initialize min-max scaler
mm_scaler = MinMaxScaler()
column_names = df.columns.tolist()
column_names.remove('Churn')

# Transform all attributes
df[column_names] = mm_scaler.fit_transform(df[column_names])
df.sort_index(inplace=True)
df.head()
```

```python
# Selecting the Feature, by remove the unused feature
feature = ['Churn']
train_feature = df.drop(feature, axis=1)

# Set The Target
train_target = df["Churn"]
```

```python
# Show the Feature
train_feature.head(5)
```

```python
# Split Data
from sklearn.model_selection import train_test_split, cross_val_score
X_train, X_test, y_train, y_test = train_test_split(train_feature ,train_target, shuffle = True, test_size=0.9, random_state=1)
```

```python
# Show the training data
X_train.head()
```

```python
# Import Library
from sklearn.neural_network import MLPClassifier

# Fitting Model
mlp = MLPClassifier(hidden_layer_sizes=(5), activation = 'relu', solver = 'adam',max_iter= 10000, verbose = True)
mlp = mlp.fit(X_train,y_train)

# Prediction to Test Dataset
y_predmlp = mlp.predict(X_test)
```

```python
print('Number of Layer =', mlp.n_layers_)
print('Number of Iteration =', mlp.n_iter_)
print('Current loss computed with the loss function =', mlp.loss_)
```

```python
# Import the metrics class
from sklearn import metrics


# Confussion Matrix
cnf_matrixmlp = metrics.confusion_matrix(y_test, y_predmlp)
cnf_matrixmlp
```

```python
# Show the Accuracy, Precision, Recall, F1, etc.
acc_mlp = metrics.accuracy_score(y_test, y_predmlp)
prec_mlp = metrics.precision_score(y_test, y_predmlp)
rec_mlp = metrics.recall_score(y_test, y_predmlp)
f1_mlp = metrics.f1_score(y_test, y_predmlp)
kappa_mlp = metrics.cohen_kappa_score(y_test, y_predmlp)

print("Accuracy:", acc_mlp)
print("Precision:", prec_mlp)
print("Recall:", rec_mlp)
print("F1 Score:", f1_mlp)
print("Cohens Kappa Score:", kappa_mlp)
```

## Q1

**Split Data:** Splitting the dataset into two parts: the training set and the test set.

- `X_train` and `X_tes`t will contain the features (independent variables) for the training and testing datasets, respectively.
- `y_train` and `y_test` will contain the corresponding target labels (the 'Churn' column in this case) for the training and testing datasets, respectively.
- `shuffle=True` ensures that the data is randomly shuffled before splitting.
- `test_size=0.3` means that 30% of the data will be used for testing, and 70% for training.
- `random_state=1` sets a seed for the random number generator to ensure reproducibility. Different values will result in different random splits

**Confussion Matrix:** A table that is often used to describe the performance of a classification model. It shows the counts of True Positives (TP), True Negatives (TN), False Positives (FP), and False Negatives (FN).

- `y_test` contains the true labels of the test dataset
- `y_predmlp` contains the predicated labels generated by the MLPClassifier

**Other definitions:**
- `accuracy_score` calculates the accuracy of the model, which is the proportion of correctly predicted instances.
- `precision_score` calculates precision, which is the ratio of true positives to the total predicted positives. It measures how many of the predicted positive cases were correct.
- `recall_score` calculates recall, which is the ratio of true positives to the total actual positives. It measures how many of the actual positive cases were correctly predicted.
- `f1_score` calculates the F1 score, which is the harmonic mean of precision and recall. It provides a balance between precision and recall.
- `cohen_kappa_score` calculates Cohen's Kappa Score, which measures the agreement between the predicted and actual labels while taking into account the possibility of agreement by chance.

```python
# Import libraries
import pandas as pd
from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_squared_error

# Load the dataset with TotalCharges as the target variable
url = 'https://raw.githubusercontent.com/rc-dbe/bigdatacertification/master/dataset/churn_trasnsformed_new.csv'
df = pd.read_csv(url, sep=',')

# Remove "Unnamed: 0" column
df = df.drop("Unnamed: 0", axis=1)

# initialize min-max scaler
mm_scaler = MinMaxScaler()
column_names = df.columns.tolist()

# Transform all attributes
df[column_names] = mm_scaler.fit_transform(df[column_names])
df.sort_index(inplace=True)

# Set TotalCharges as the target variable
target = 'TotalCharges'
feature_cols = [col for col in df.columns if col != target]

train_feature = df[feature_cols]
train_target = df[[target]]

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(train_feature, train_target, test_size=0.3, random_state=1)

# Train a Linear Regression model
regressor = GradientBoostingRegressor()
model = MultiOutputRegressor(regressor)
model.fit(x_train, y_train)

# Prediction to Test Dataset
y_predict = model.predict(x_test)

# Calculate the mean squared error (MSE)
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error (MSE):", mse)
```
